```python
import pandas as pd
df=pd.read_csv("C:\\Users\\abc\\Desktop\\random walk.csv")
df
```

```python
x=list(df["t"])
x
```

```python
y=list(df["starting pt"])
y
```

```python
import math 
import random
import numpy as np
li=[]
for i in range(len(x)):
    n = random.random()
    print(n)
    li.append(n)
    
li
```

```python
df['probability']=li
df
```

```python
import numpy as np
li1=[]
for i in range(len(x)):
    s = np.random.normal(0, 1.2247)
    print(s)
    li1.append(s)
    
li1
```

```python
df['random steps']=li1
df
```

```python
a=np.cumsum(df["random steps"])
a
```

```python
df['cumulative random steps']=a
df
```

```python
b=list(df["cumulative random steps"])
b
```

```python
y=list(df["starting pt"])
y
```

```python
import math
import numpy as np
li2=[]
for i in range(len(x)):
    c=(y[i]+b[i])
    print(c)
    li2.append(c)
    
li2
```

```python
df['current step']=li2
df
```

```python
r=list(df["random steps"])
r
```

```python
import math
import numpy as np
li3=[]
for i in range(len(x)-1):
    d=(r[i+1]-r[i])
    print(d)
    li3.append(d)
    
li3
```

```python
df2=pd.DataFrame(li3)
df2
```

```python
import matplotlib.pyplot as plt
plt.plot(df2,marker="o")
plt.title("Random walk")
plt.show
  
```
